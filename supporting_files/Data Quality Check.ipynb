## Load all json files

```python
import gzip
import json
import pandas as pd

#load brands data
brands = []
with gzip.open('brands.json.gz', 'rt', encoding='utf-8') as f:
    for line in f:
        try:
            json_obj = json.loads(line)
            brands.append(json_obj)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
brands_df = pd.DataFrame(brands)

#load receipts data
receipts = []
with gzip.open('receipts.json.gz', 'rt', encoding='utf-8') as f:
    for line in f:
        try:
            json_obj = json.loads(line)
            receipts.append(json_obj)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
receipts_df = pd.DataFrame(receipts)
            
#load users data
users = []
with gzip.open('users.json.gz', 'rt', encoding='utf-8') as f:
    for line in f:
        try:
            json_obj = json.loads(line)
            users.append(json_obj)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
users_df = pd.DataFrame(users)
```
